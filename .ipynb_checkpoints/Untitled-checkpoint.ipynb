```python
def func(*aList):
    col = len(aList)
    row = 0
    for i in aList:
        if len(i) > row:
            row = len(i)
    majors = aList
    
    locs = ['right','bottom','top','left']
    
    fig, ax = plt.subplots(row,col, figsize=(8,12))
    for k in range(len(majors)):
        for i in range(row):
            ax[i,k].plot(women_degrees.Year, women_degrees[majors[k][i]])
            ax[i,k].plot(women_degrees.Year, 100-women_degrees[majors[k][i]])
            ax[i,k].set_title(majors[k][i])
            
            ax[i,k].set_xlim(1968,2011)
            ax[i,k].set_ylim(0,100)
            ax[i,k].set_yticks([1,100])
            
            ax[i,k].axhline(50)
            
            for s in locs:
                ax[i,k].spines[s].set_visible(False)
            
            
    
            
    
  


            
```

```python
func(stem,other)
```
