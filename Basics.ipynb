# Guided Project: Visualizing The Gender Gap in College Degrees

In this guided project, we'll extend the work we did in the last two missions on visualizing the gender gap across college degrees. So far, we mostly focused on the STEM degrees but now we will generate line charts to compare across all degree categories. In the last step of this guided project, we'll explore how to export the final diagram we create as an image file.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_grey = (171/255, 171/255, 171/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology',
             'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism',
                 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration',
              'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig, axes = plt.subplots(6, 3, figsize=(18,15))
fig.delaxes(axes[5][1])

locs = ['right','bottom','top','left']

for i in range(len(axes)):
    axes[i][0].plot(women_degrees['Year'], 100-women_degrees[stem_cats[i]], c=cb_orange, label='Men', linewidth=3)
    axes[i][0].plot(women_degrees['Year'], women_degrees[stem_cats[i]], c=cb_dark_blue, label='Women', linewidth=3)
    axes[i][0].set_title(stem_cats[i])
    for s in locs:
        axes[i][0].spines[s].set_visible(False)
    axes[i][0].set_xlim(1968, 2011)
    axes[i][0].set_ylim(0,100)
    axes[i][0].set_yticks([1,100])
    
    axes[i][0].axhline(50)
    axes[i][0].axhline(50, c=cb_grey, alpha=0.3)
    
    axes[i][0].tick_params(bottom=False, top=False, left=False, right=False,labelbottom=False)
    
    if i == 0:
        axes[i][0].text(2005, 87, 'Men')
        axes[i][0].text(2002, 8, 'Women')
    elif i == 5:
        axes[i][0].text(2005, 62, 'Men')
        axes[i][0].text(2001, 35, 'Women')
        axes[i][0].tick_params(labelbottom=True)

for i in range(len(axes)-1):
    axes[i][1].plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[i]], c=cb_orange, label='Men', linewidth=3)
    axes[i][1].plot(women_degrees['Year'], women_degrees[lib_arts_cats[i]], c=cb_dark_blue, label='Women', linewidth=3)
    axes[i][1].set_title(lib_arts_cats[i])
    for s in locs:
        axes[i][1].spines[s].set_visible(False)
    axes[i][1].set_xlim(1968, 2011)
    axes[i][1].set_ylim(0,100)
    axes[i][1].set_yticks([1,100])
    
    axes[i][1].axhline(50)
    axes[i][1].axhline(50, c=cb_grey, alpha=0.3)
    
    axes[i][1].tick_params(bottom=False, top=False, left=False, right=False,labelbottom=False)
    
    if i == 0:
        axes[i][1].text(2005, 87, 'Men')
        axes[i][1].text(2002, 8, 'Women')
    elif i == 4:
        axes[i][1].text(2005, 62, 'Men')
        axes[i][1].text(2001, 35, 'Women')
        axes[i][1].tick_params(labelbottom=True)
    
for i in range(len(axes)):
    axes[i][2].plot(women_degrees['Year'], 100-women_degrees[other_cats[i]], c=cb_orange, label='Men', linewidth=3)
    axes[i][2].plot(women_degrees['Year'], women_degrees[other_cats[i]], c=cb_dark_blue, label='Women', linewidth=3)
    axes[i][2].set_title(other_cats[i])
    for s in locs:
        axes[i][2].spines[s].set_visible(False)
    axes[i][2].set_xlim(1968, 2011)
    axes[i][2].set_ylim(0,100)
    axes[i][2].set_yticks([1,100])
    
    axes[i][2].axhline(50)
    axes[i][2].axhline(50, c=cb_grey, alpha=0.3)
    
    axes[i][2].tick_params(bottom=False, top=False, left=False, right=False,labelbottom=False)
    
    if i == 0:
        axes[i][2].text(2005, 87, 'Men')
        axes[i][2].text(2002, 8, 'Women')
    elif i == 5:
        axes[i][2].text(2005, 62, 'Men')
        axes[i][2].text(2001, 35, 'Women')
        axes[i][2].tick_params(labelbottom=True)

#fig.savefig('degree_genders.svg')
```
