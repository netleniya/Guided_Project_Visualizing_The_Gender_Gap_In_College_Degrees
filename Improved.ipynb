# Guided Project: Visualizing The Gender Gap in College Degrees

In this guided project, we'll extend the work we did in the last two missions on visualizing the gender gap across college degrees. So far, we mostly focused on the STEM degrees but now we will generate line charts to compare across all degree categories. In the last step of this guided project, we'll explore how to export the final diagram we create as an image file.


A `plotter()` function that takes an arbitrary number of lists was implmented. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stem = [
    'Engineering', 'Computer Science', 'Psychology',
    'Biology', 'Physical Sciences', 'Math and Statistics'
]
arts = [
    'Foreign Languages', 'English', 'Communications and Journalism',
    'Art and Performance', 'Social Sciences and History'
]
other= [
    'Health Professions', 'Public Administration',
    'Education', 'Agriculture','Business', 'Architecture'
]
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
def plotter(*list_args):
    """Generates line charts for given input and saves an image
    
    Usage:
    -----
    plotter(list1, list2,...,listN)
    
    Parameters:
    ----------
    *list_args : arbitrary number of list_args > 0
    
    Returns :
    --------
    None
    
    Raises:
    -------
    ValueError
        If no arguments are given
    
    """
        
    if len(list_args) == 0:
        raise ValueError("Number of columns must be a positive integer")
    
    col = len(list_args)
    row = 1
    for i in list_args:
        if len(i) > row:
            row = len(i)
    
    majors = list_args
    locs = ['right','bottom','top','left']
    wor = row
        
    fig, ax = plt.subplots(row,col, figsize=(12,18))
    
    if col == 1:
        ax = np.atleast_2d(ax).reshape(-1,1)
    
    for k in range(len(majors)):
        if len(majors[k]) != row:
            row = len(majors[k])
                   
        for i in range(row):
            ax[i,k].plot(women_degrees.Year, women_degrees[majors[k][i]], c='tab:blue')
            ax[i,k].plot(women_degrees.Year, 100-women_degrees[majors[k][i]], c='tab:orange')
            ax[i,k].set_title(majors[k][i])

            ax[i,k].set_xlim(1968,2011)
            ax[i,k].set_ylim(0,100)
            ax[i,k].set_yticks([1,100])
            ax[i,k].axhline(50, c='tab:gray')
            ax[i,k].tick_params(bottom=False, top=False, left=False, right=False,labelbottom=False)

            for s in locs:
                ax[i,k].spines[s].set_visible(False)

            if i == 0:
                ax[i,k].text(2005, 87, 'Men')
                ax[i,k].text(2002, 8, 'Women')
            elif i == (row-1):
                ax[i,k].text(2005, 62, 'Men')
                ax[i,k].text(2001, 35, 'Women')
                ax[i,k].tick_params(labelbottom=True)
                
                if row != wor:
                    ax[-1,k].set_visible(False)
                                
    fig.savefig('degree_genders.svg')
```

```python
plotter(stem,arts,other)
```
